```python
#I used text8 dataset for the training: https://mattmahoney.net/dc/textdata.html

with open("text8", "r") as f:
    text = f.readline().lower()
    
print(f"Total number of symbols: {len(text)}, Total number of words: {len(text.split(' '))}, Total number of unique words: {len(set(text.split(' ')))}")
```

```python
text = ' '.join(text.split(' ')[: 10**6])

print(f"Total number of words: {len(text.split(' '))}, Total number of unique words: {len(set(text.split(' ')))}")
```

```python
from w2v import Word2Vec

model = Word2Vec(text)
```

```python
model.fit()
```

```python
model.most_similar("city")
```
